### Load user profile from JSON

```python
import json

with open("user.json", "r") as file:
    data = json.load(file)

print(data)
```

### Save student data

```python
with open("student.json", 'w') as file:
    json.dump(data, file, indent=4)
```

### Convert dictionary to JSON

**Method 1**: *Convert to a JSON String. Use this when you need the JSON representation as a string variable (e.g., for APIs or logging)*

```python
import json

data_dict = {"name": "maiyan", "age": 22, "city": "Dhaka"}

data = json.dumps(data_dict)
print(data)
```

**Method 2:** *Save Directly to a JSON File. Use this when you want to create or overwrite a static physical file.*

```python
import json

data_dict = {"name": "maiyan", "age": 22, "city": "Dhaka"}

with open("dict_to_json.json", 'a') as file:
    json.dump(data_dict, file, indent=4)
```
