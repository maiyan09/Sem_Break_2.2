```python
from bs4 import BeautifulSoup

with open("sample.html", 'r') as file:
    html_doc = file.read()

soup = BeautifulSoup(html_doc, 'html.parser')
print(soup.prettify())
```

### Title

```python
soup.title
```

```python
soup.title.name
```

```python
soup.title.string
```

```python
soup.title.parent.name
```

## Return first p tag

```python
print(soup.p)
```

### Return class name

```python
print(soup.p["class"])
```

## Return all a tag as a list

```python
print(soup.find_all('a'))
```

## Find any tag by id
find return only one tag but find_all return all the tag with same is

```python
print(soup.find_all(id = 'link3'))
```

##  Extracting all the URLs found within a page’s <a> tags:

```python
for i in soup.find_all('a'):
    print(i.get('href'))
```

## Extracting all the text from a page

```python
page = soup.get_text()
print(page)
```
