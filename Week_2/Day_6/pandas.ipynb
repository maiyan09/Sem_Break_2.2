### Create DataFrame

```python
import pandas as pd

data = {
    "Name" : ["Maiyan", "Messi", "Neymer"],
    "Age" : [23, 39, 36],
    "Country" : ["Bangladesh", "Argentina", "Brazil"]
}
dataFrame = pd.DataFrame(data)
print(dataFrame)
```

### Named Indexes
With the index argument, We can name your own indexes.

```python
dataFrame = pd.DataFrame(data, index=[9, 10, 11]) #Jersey number 😂
print(dataFrame)
```

## Read CSV

```python
csvData = pd.read_csv("students.csv")
print(csvData)
# print(csvData.head()) Shows first 5 rows
# print(csvData.tail()) Shows last 5 rows
# print(df.to_string()) print the entire DataFrame.
```

### Colums name

```python
print(csvData.columns)
```

### Filter Data

Print student name who's age is greater than 16

```python
age16 = csvData[csvData["Age"]>16]
print(age16)
```

Age < 16 and Class = 9

```python
age16_class10 = csvData[(csvData["Age"] < 16) & (csvData["Class"] == 9)]
print(age16_class10)
```

Average marks of all subject 

```python
# csvData.groupby(["Class"]).mean(numeric_only=True)
avg = csvData.groupby("Class")[["Physics","Math","Chemistry"]].mean()
print(avg)
```

Sort number in descending order

```python
sortNumber = csvData.sort_values(by=["Physics", "Math", "Chemistry"], ascending=False)
print(sortNumber)
```

Made chnages on main file

```python
csvData.sort_values(by=["Physics", "Math", "Chemistry"], ascending=False, inplace=True)
print(csvData)
```

### Export CSV

Removes extra index column (index=False)

```python
csvData.to_csv("students_output.csv", index=False)
```
