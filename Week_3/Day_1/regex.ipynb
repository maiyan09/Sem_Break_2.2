### ```Remove numbers```

````python
import re
````

re.sub() is used to replace parts of a string that match a regular expression.
## <i> Syntax </i> 
### ```re.sub(pattern, replacement, string)```
```pattern``` → What you want to find (using regex).<br>
```replacement``` → What you want to replace it with.<br>
```string``` → The original text.<br>
It returns a new string. The original string is unchanged.

````python
text = "Hello123. My 34 name8 is 34 Samiur 11 Rahman 12300 Maiyan."

pattern = r"\d"

text = re.sub(pattern, "", text)
print(text)
````

### ``` Extract emails ```

````python
text = '''
Please read the following message carefully and extract all email addresses from it. Students should identify
valid emails such as alice@example.com and bob.smith@school.edu within the text. Some messages may also
include support@company.org or info@service.net for practice purposes. Try to separate personal and 
professional emails and list them clearly after reading. Do not miss any address even if it appears inside
a sentence or followed by punctuation marks. Also check variations like admin@portal.io team.hr@workplace.co 
and training@learn.org for accuracy. 
'''

pattern = r"[\w.-]+@[\w.-]+\.\w+"
result = re.findall(pattern, text)

print(result)
````

### ```Count words```

````python
text = '''
Lorem Ipsum is simply dummy text of the printing and typesetting industry. Lorem Ipsum has been the
industry's standard dummy text ever since 1966, when designers at Letraset and James Mosley, the librarian
at St Bride Printing Library in London, took a 1914 Cicero translation and scrambled it to make dummy text
for Letraset's Body Type sheets. It has survived not only many decades, but also the leap into electronic
typesetting, remaining essentially unchanged. It was popularised thanks to these sheets and more recently
with desktop publishing software like Aldus PageMaker and Microsoft Word including versions of Lorem Ipsum.
'''
pattern  = r"\b\w+(?:'\w+)?\b"
words = re.findall(pattern, text)

print(f"Number of words: {len(words)}")
````

### ```Remove spaces```

````python
text = '''
  This's    is   a     simple      dummy   text   with   irregular   spaces.

It      is      created    for    testing      word     counting   and    regex   patterns.

Some     words      are      separated       by       multiple        spaces       intentionally.

Use     this      text      to      practice      splitting      and      normalization      tasks.

End      of      sample        text.
'''
pattern = r'\s+'
text = re.sub(pattern, ' ', text)
print(text)
````
