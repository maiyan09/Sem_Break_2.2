````python
import asyncio
````

```asyncio``` helps Python:

- wait without blocking
- run multiple tasks
- work asynchronously

Without this line,

``` await asyncio.sleep(3) ```

would give an error because Python wouldn't know what asyncio is.

````python
async def timer():
    print("Time started...")
    await asyncio.sleep(3)
    
    print("Halfway...")
    await asyncio.sleep(2)
    
    print("Time's up...")
````

```async``` tells Python:

"This function may need to wait for something."

Inside an async function, we can use await.

```asyncio.sleep(3)```

This means

"Wait for 3 seconds."

````python
# asyncio.run(timer()) > for .py file
await timer() # For .ipynb file
````

### Key Concept
| Keyword            | Meaning                                                             |
| ------------------ | ------------------------------------------------------------------- |
| `import asyncio`   | Load Python's asynchronous programming module.                      |
| `async`            | Marks a function as asynchronous.                                   |
| `def`              | Defines (creates) a function.                                       |
| `await`            | Pause the current async function until an async operation finishes. |
| `asyncio.sleep(3)` | Wait asynchronously for 3 seconds.                                  |
| `asyncio.run()`    | Starts the async program and runs the async function.               |
