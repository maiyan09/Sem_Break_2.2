```python
import asyncio

async def cooking():
    print("Cooking Started..")
    await asyncio.sleep(2)
    print("Cooking Finished..")
    
async def cleaning():
    print("Cleaning Started..")
    await asyncio.sleep(4)
    print("Cleaning Finished..")
    
async def studying():
    print("Studying Started..")
    await asyncio.sleep(3)
    print("Studying Finished..")
    
async def main():
    await asyncio.gather(
        cooking(),
        cleaning(),
        studying()
    )
    
# asyncio.run(main())
await main()
```
